```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

import seaborn as sns
```

# Data Visualisation using PCA for Bag of Words
1. Easy (inter industry)
2. Hard (tech industry)

```python
bag_easy = pd.read_csv("bag_easy_train.csv")
bag_easy_labels = bag_easy["category"]
bag_easy = bag_easy.drop(labels="category", axis=1)

bag_difficult = pd.read_csv("bag_difficult_train.csv")
bag_difficult_labels = bag_difficult["category"]
bag_difficult = bag_difficult.drop(labels="category", axis=1)
```

```python
pca1 = PCA(n_components=2)
pca2 = PCA(n_components=2)
```

```python
pca1.fit(bag_easy)
bag_easy_pca = pca1.transform(bag_easy)
bag_easy_pca = pd.DataFrame(bag_easy_pca, columns=["feature_1", "feature_2"])
bag_easy_pca["category"] = bag_easy_labels
```

```python
pca2.fit(bag_difficult)
bag_difficult_pca = pca2.transform(bag_difficult)
bag_difficult_pca = pd.DataFrame(bag_difficult_pca, columns=["feature_1", "feature_2"])
bag_difficult_pca["category"] = bag_difficult_labels
```

```python
sns.relplot(data=bag_easy_pca, x="feature_1", y="feature_2", hue="category", kind="scatter", height=6)
```

```python
sns.relplot(data=bag_difficult_pca, x="feature_1", y="feature_2", hue="category", kind="scatter", height=6)
```
