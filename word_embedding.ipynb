```python
import re
import nltk
from nltk.corpus import stopwords
from sentence_transformers import SentenceTransformer
import pandas as pd
```

```python
is_working_with_easy_dataset = False
```

```python
def remove_html(txt):
    txt = txt.lower()
    txt = txt.split("Job Type")
    txt = txt[0]
    txt = re.findall("\>(.*?)\<", txt)
    txt = " ".join(txt)
    txt = txt.split()
    stop_words = stopwords.words("english")
    txt = [x for x in txt if x not in stop_words]
    txt = " ".join(txt)
    return txt

def embed(model, txt):
    txt = [txt]
    embeddings = model.encode(txt)
    for sentence, embedding in zip(txt, embeddings):
        return embedding
```

```python
if is_working_with_easy_dataset:
    input_train_filename = "jd_easy_train.csv"
    input_test_filename = "jd_difficult_train.csv"
else:
    input_train_filename = "jd_difficult_train.csv"
    input_test_filename = "jd_difficult_test.csv"

jd_train = pd.read_csv(input_train_filename, keep_default_na=False)
jd_test = pd.read_csv(input_test_filename, keep_default_na=False)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embedding_vectors_train = list(pd.Series.tolist(jd_train.description.apply(lambda x: embed(model, remove_html(x)))))
embedding_train = pd.concat([jd_train.category, pd.DataFrame(embedding_vectors_train)], axis=1)
embedding_train
```

```python
embedding_vectors_test = list(pd.Series.tolist(jd_test.description.apply(lambda x: embed(model, remove_html(x)))))
embedding_test = pd.concat([jd_test.category, pd.DataFrame(embedding_vectors_test)], axis=1)
embedding_test
```

```python
if is_working_with_easy_dataset:
    output_train_filename = "embedding_easy_train.csv"
    output_test_filename = "embedding_easy_test.csv"
else:
    output_train_filename = "embedding_difficult_train.csv"
    output_test_filename = "embedding_difficult_test.csv"

embedding_train.to_csv(output_train_filename, index=False)
embedding_test.to_csv(output_test_filename, index=False)
```
