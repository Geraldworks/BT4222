```python
import pandas as pd
import modelEnsemble as me
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
```

### Import the Models

```python
logistic_regression_model = LogisticRegression(random_state = 0)
svm_model = SVC(random_state = 0, kernel = "poly", probability = True, C=10, gamma=1)
adaboost_model = AdaBoostClassifier(random_state = 0)
gradient_boosting_model = GradientBoostingClassifier(random_state = 0)
gaussiannb_model = GaussianNB()
decision_tree_model = DecisionTreeClassifier(random_state = 0)
random_forest_model = RandomForestClassifier(random_state = 0, max_depth=8, n_estimators=400)
xgboost_model = xgb.XGBClassifier(random_state = 0, max_depth=4, n_estimators=700)

estimators = [('LogisticClassifier', logistic_regression_model), ('SVM', svm_model), ('AdaBoosting', adaboost_model), ('Gradient Boosting', gradient_boosting_model), ('Gaussian NB', gaussiannb_model), ('Decision Tree', decision_tree_model), 
                ('Randome Forest', random_forest_model), ('XGBoost', xgboost_model)]
```

### Traditional Models Embedding Difficult

```python
embeddingDifficultReport = me.OnehotReportGenerator("embedding_difficult_train.csv", "embedding_difficult_test.csv")
```

```python
embeddingDifficultReport.runBasicModels(estimators)
```

```python
embeddingDifficultReport.makeReport()
```
